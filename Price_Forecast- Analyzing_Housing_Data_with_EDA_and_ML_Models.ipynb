```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
x = df1.drop(['Suburb','Address','Method','SellerG','CouncilArea','Regionname','Propertycount','ParkingArea','Price','Bathroom'],axis=1)
y = df1['Price']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
```

# LinearRegression

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(
    fit_intercept=True,
    n_jobs=None
)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mlr=mean_squared_error(y_test, y_pred)
print('mean_squared_error is :',mlr)

malr=mean_absolute_error(y_test, y_pred)
print('mean_absolute_error is :',malr)

rml=r2_score(y_test, y_pred)
print('r2_score is :',rml)
```
